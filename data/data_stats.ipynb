```python
import pandas as pd

# Load the dataset
file_path = '/data1/bubble3jh/cluster-regression/data/data_cut_0.csv'
df = pd.read_csv(file_path)
# Step 1: Exclude the specified categorical columns
categorical_columns = ['gender', 'is_korean', 'primary case', 'job_idx', 'rep_idx', 'place_idx', 'add_idx']
df_numerical = df.drop(columns=categorical_columns)

# Step 2: Group by the 'cluster' column and compute the mean of numerical columns for each cluster
cluster_means = df_numerical.groupby('cluster').mean().reset_index()

# Step 3: Group by 'cut_date' and count the number of rows in each group in the transformed DataFrame
cut_date_counts = cluster_means.groupby('cut_date').size().reset_index(name='count')

# Display the resulting DataFrame
cut_date_counts.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Create the bar plot
plt.figure(figsize=(12, 8))

# Use seaborn for a more professional look
sns.barplot(x='cut_date', y='count', data=cut_date_counts, palette='viridis')

# Adding titles and labels with appropriate font sizes
plt.xlabel('Cut Off', fontsize=20)
plt.ylabel('Count', fontsize=20)

# Customize ticks
print(cut_date_counts['cut_date'])
plt.xticks(ticks=[0,1,2,3,4], labels=[int(x) for x in cut_date_counts['cut_date']], fontsize=20)
plt.yticks(fontsize=20)

# Remove grid lines
plt.grid(False)

# Remove the top and right spines for a cleaner look
sns.despine(trim=True)

# Ensure data values are displayed as integers and add value labels on top of the bars
# Ensure data values are displayed as integers and add value labels on top of the bars with a small offset
for index, row in cut_date_counts.iterrows():
    plt.text(index, row['count'] + 10, int(row['count']), color='black', ha="center", fontsize=20)


# # Save the plot as a high-resolution image
output_plot_path = '/data1/bubble3jh/cluster-regression/data/figures/cutoff_counts.png'
plt.savefig(output_plot_path, dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = './data_task.csv'
df = pd.read_csv(file_path)

# Count the values for each specified column
job_idx_counts = df['job_idx'].value_counts()
rep_idx_counts = df['rep_idx'].value_counts()
place_idx_counts = df['place_idx'].value_counts()
add_idx_counts = df['add_idx'].value_counts()

# Set the style for the plot
sns.set(style="whitegrid")

# Function to create and save individual bar plots
def create_bar_plot(data, title_name, xlabel_name, output_path, is_red=False):
    plt.figure(figsize=(12, 8))
    ax = sns.barplot(x=data.index, y=data.values, color='darkgray')
    plt.title(f'', fontsize=20)
    plt.xlabel('', fontsize=16)
    plt.ylabel('', fontsize=16)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=24)
    print(data.index)
    if is_red:
        for i, bar in enumerate(ax.patches):
            if data.index[i] == 6:
                bar.set_color('lightcoral')
    plt.tight_layout()
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.show()

# Create and save plots for each column with customized names
create_bar_plot(job_idx_counts, 'Job Idx', 'Job Idx', './job_idx_distribution.png')
create_bar_plot(rep_idx_counts, 'Report Idx', 'Report Idx', './rep_idx_distribution.png')
create_bar_plot(place_idx_counts, 'Place Idx', 'Place Idx', './place_idx_distribution.png')
create_bar_plot(add_idx_counts, 'Address Idx', 'Address Idx', './add_idx_distribution.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터셋 로드
file_path = './data_task.csv'
df = pd.read_csv(file_path)

# 값 매핑
df['gender'] = df['gender'].map({1: 'Male', 0: 'Female'})
df['is_korean'] = df['is_korean'].map({1: 'Korean', 0: 'Foreigner'})

# 이진 범주형 데이터에 대한 파이 차트를 생성하고 저장하는 함수
def create_pie_chart(data, column_name, title_name, output_path):
    plt.figure(figsize=(8, 8))
    pie_chart = data[column_name].value_counts().plot.pie(
    autopct='%1.1f%%', startangle=90, colors=['lightgray', 'darkgray'], textprops={'fontsize': 28}
    )
    plt.title('', fontsize=20)
    plt.ylabel('')

    # 레이블 폰트 크기 설정
    for text in pie_chart.texts:
        text.set_fontsize(28)  # 레이블 폰트 크기 설정

    plt.tight_layout()
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.show()

# 'gender' 및 'is_korean'에 대한 파이 차트를 생성하고 저장
create_pie_chart(df, 'gender', 'Gender Distribution', './gender_distribution_pie.png')
create_pie_chart(df, 'is_korean', 'Korean Distribution', './is_korean_distribution_pie.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
file_path = './data_cut_0.csv'
df = pd.read_csv(file_path)

# Set the style for the plot
sns.set(style="whitegrid")

# Function to create bins based on unique values
def create_bins(data, num_bins):
    unique_values = np.unique(data)
    if len(unique_values) <= num_bins:
        return unique_values
    else:
        return np.linspace(min(unique_values), max(unique_values), num_bins)

# Get the unique cut_dates
cut_dates = df['cut_date'].unique()

# Create plots for each cut_date
for cut_date in cut_dates:
    # Filter data for the current cut_date
    df_cut = df[df['cut_date'] == cut_date]
    
    # Compute the mean for each cluster
    categorical_columns = ['gender', 'is_korean', 'primary case', 'job_idx', 'rep_idx', 'place_idx', 'add_idx']
    df_numerical = df_cut.drop(columns=categorical_columns)
    cluster_means = df_numerical.groupby('cluster').mean().reset_index()
    
    # Create bins for 'y' and 'd' in the clustered DataFrame
    y_bins = create_bins(cluster_means['y'], 30)
    d_bins = create_bins(cluster_means['d'], 30)
    
    # Create a single plot for 'y' and 'd' distributions in the clustered DataFrame
    plt.figure(figsize=(12, 8))
        
    # Plot distribution for 'y'
    sns.histplot(cluster_means['y'], bins=y_bins, kde=True, color='skyblue', label='Confirmed Case', alpha=0.6)

    # Plot distribution for 'd'
    sns.histplot(cluster_means['d'], bins=d_bins, kde=True, color='lightcoral', label='Cluster Duration', alpha=0.6)

    # Adding titles and labels
    # plt.title(f'Distribution of y and d for cut_date {cut_date}', fontsize=20)
    plt.xlabel('Value', fontsize=16)
    plt.ylabel('Frequency', fontsize=16)
    plt.legend(fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    
    # Adjust layout and save the plot
    plt.tight_layout()
    output_plot_path = f'./figures/y_d_distribution_cut_date_{cut_date}.png'
    plt.savefig(output_plot_path, dpi=300, bbox_inches='tight')
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 데이터셋 로드
file_path = './data_cut_0.csv'
df = pd.read_csv(file_path)
# Step 1: Exclude the specified categorical columns
categorical_columns = ['gender', 'is_korean', 'primary case', 'job_idx', 'rep_idx', 'place_idx', 'add_idx']
df_numerical = df.drop(columns=categorical_columns)

# Step 2: Group by the 'cluster' column and compute the mean of numerical columns for each cluster
cluster_means = df_numerical.groupby('cluster').mean().reset_index()
# 스타일 설정
sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

# 데이터 변환: cut_date를 문자열로 변환하여 리지라인 차트의 구분자로 사용
cluster_means['cut_date_str'] = cluster_means['cut_date'].astype(str)

# 색상 팔레트 생성
pal = sns.color_palette(palette='coolwarm', n_colors=cluster_means['cut_date_str'].nunique())

# y 변수에 대한 리지라인 차트 생성
g = sns.FacetGrid(cluster_means, row="cut_date_str", hue="cut_date_str", aspect=15, height=0.75, palette=pal)
bw_adjust=0.3
# KDE 플롯 추가
g.map(sns.kdeplot, 'y', bw_adjust=bw_adjust, clip_on=False, fill=True, alpha=1, linewidth=1.5)
g.map(sns.kdeplot, 'y', bw_adjust=bw_adjust, clip_on=False, color="w", lw=2)
g.map(plt.axhline, y=0, lw=2, clip_on=False)

# cut_date를 텍스트로 추가
for i, ax in enumerate(g.axes.flat):
    ax.text(-0.05, 0.02, cluster_means['cut_date_str'].unique()[i],
            fontweight='bold', fontsize=15,
            color=ax.lines[-1].get_color())

# 서브플롯 겹치게 조정
g.fig.subplots_adjust(hspace=-0.3)

# 축 제목, 눈금, 스파인 제거
g.set_titles("")
g.set(yticks=[])
g.despine(bottom=True, left=True)

# 축 라벨 설정
for ax in g.axes.flat:
    ax.set_xlabel('Value of y', fontsize=15, fontweight='bold')
    ax.set_ylabel('Count', fontsize=15, fontweight='bold')
    ax.tick_params(axis='x', labelsize=12)
    ax.tick_params(axis='y', labelsize=12)

g.fig.suptitle('Distribution of y by Cut Date', ha='right', fontsize=20, fontweight='bold')

plt.show()

# d 변수에 대한 리지라인 차트 생성
g = sns.FacetGrid(cluster_means, row="cut_date_str", hue="cut_date_str", aspect=15, height=0.75, palette=pal)

# KDE 플롯 추가
g.map(sns.kdeplot, 'd', bw_adjust=bw_adjust, clip_on=False, fill=True, alpha=1, linewidth=1.5)
g.map(sns.kdeplot, 'd', bw_adjust=bw_adjust, clip_on=False, color="w", lw=2)
g.map(plt.axhline, y=0, lw=2, clip_on=False)

# cut_date를 텍스트로 추가
for i, ax in enumerate(g.axes.flat):
    ax.text(-0.05, 0.02, cluster_means['cut_date_str'].unique()[i],
            fontweight='bold', fontsize=15,
            color=ax.lines[-1].get_color())

# 서브플롯 겹치게 조정
g.fig.subplots_adjust(hspace=-0.3)

# 축 제목, 눈금, 스파인 제거
g.set_titles("")
g.set(yticks=[])
g.despine(bottom=True, left=True)

# 축 라벨 설정
for ax in g.axes.flat:
    ax.set_xlabel('Cluster Duration', fontsize=22)
    ax.set_ylabel('', fontsize=15, fontweight='bold')
    ax.tick_params(axis='x', labelsize=12)
    ax.tick_params(axis='y', labelsize=12)

# g.fig.suptitle('Distribution of d by Cut Date', ha='right', fontsize=20, fontweight='bold')

plt.show()
```

```python
# Generating an adjusted lollipop plot with extended y-axis limits
import matplotlib.pyplot as plt
import numpy as np

# 데이터 설정
# Updated models and values
models = ['Linear', 'Ridge', 'MLP', 'Transformer', 'iTransformer', 'TARNet', 'Dragonnet', 'CEVAE', 'CEVT']
values = [-7.674, -7.214, -7.631, -2.912, -0.144, -23.299, -24.576, -6.170, 2.891]
ground_truth = 1.789

# Sorting models and values by values in descending order
sorted_indices = np.argsort(values)[::-1]
models = [models[i] for i in sorted_indices]
values = [values[i] for i in sorted_indices]

# 그래프 그리기
plt.figure(figsize=(12, 6), dpi=300)

# Lollipop plot
for i, value in enumerate(values):
    plt.plot([i, i], [0, value], color='gray', linewidth=2)
    plt.plot(i, value, 'o', color='blue' if value >= 0 else 'red', markersize=12)

# Ground Truth line
plt.axhline(ground_truth, color='green', linestyle='--', linewidth=1.5, label='Ground Truth')
plt.axhline(0, color='black', linewidth=0.8)

# Ground Truth text annotation
plt.text(len(models) - 1.4, ground_truth + 0.3, f'{ground_truth:.2f}', color='green', ha='left', va='bottom', fontsize=20)

# 점 위에 값 표시 (CEVT는 약간 위로, Dragonnet은 약간 아래로 이동)
for i, value in enumerate(values):
    if models[i] == 'CEVT':
        plt.text(i, value + 0.5, f'{value:.2f}', ha='center', va='bottom', fontsize=18)
    elif models[i] == 'Dragonnet':
        plt.text(i, value - 2, f'{value:.2f}', ha='center', va='top', fontsize=18)
    else:
        plt.text(i, value - 1, f'{value:.2f}', ha='center', va='bottom' if value > ground_truth else 'top', fontsize=18)

# 제목과 축 레이블 설정
# plt.title('Model Predictions with Ground Truth', fontsize=22)
plt.xlabel('', fontsize=20)
plt.ylabel('ITE', fontsize=23)
plt.xticks(ticks=np.arange(len(models)), labels=models, rotation=45, fontsize=18)
plt.yticks(fontsize=18)
plt.ylim(-30, 7)  # Adjusted y-axis limit to ensure Dragonnet doesn't go out of bounds
plt.legend(fontsize=20, loc='lower left')

# 그래프 저장
plt.savefig('./figures/syn_ite.png')

plt.show()
```
